```python
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = '📸 Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getTracks().forEach(track => track.stop());
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = io.BytesIO(base64.b64decode(data.split(',')[1]))
    with open(filename, 'wb') as f:
        f.write(binary.getbuffer())
    return filename
```

```python
import cv2
import numpy as np
import time
import sys

img = cv2.imread("photo.jpg")
reference_img = cv2.imread(image_path)
reference_gray = cv2.cvtColor(reference_img, cv2.COLOR_BGR2GRAY)
reference_gray = cv2.GaussianBlur(reference_gray, (21, 21), 0)

cap = cv2.VideoCapture(0)

motion_levels = []
motion_text = "Stop"

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (21, 21), 0)

    frame_delta = cv2.absdiff(reference_gray, gray)
    thresh = cv2.threshold(frame_delta, 25, 255, cv2.THRESH_BINARY)[1]
    thresh = cv2.dilate(thresh, None, iterations=2)

    contours, _ = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    motion_area = 0

    for contour in contours:
        area = cv2.contourArea(contour)
        if area < 1000:
            continue
        motion_area += area
        (x, y, w, h) = cv2.boundingRect(contour)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    motion_levels.append(motion_area)
    if len(motion_levels) > 10:
        motion_levels.pop(0)

    avg_motion = sum(motion_levels) / len(motion_levels)

    if avg_motion > 30000:
        motion_text = "Running"
    elif avg_motion > 8000:
        motion_text = "Walking"
    elif avg_motion > 1000:
        motion_text = "Moving"
    else:
        motion_text = "Stop"

    print(f"Motion Status: {motion_text}")
    sys.stdout.flush()

    cv2.putText(frame, f"Status: {motion_text}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

    output_frame = frame.copy()
    cv2.putText(output_frame, f"Status: {motion_text}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

    cv2.imshow("Motion Detection", output_frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('s'):
        cv2.imwrite("motion_status_output.jpg", output_frame)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
animals = ['dog', 'cat', 'bird', 'horse', 'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe']
plants = ['potted plant']
insects = ['butterfly']
people = ['person']
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)


def run_detection(image_path):
    results = model(image_path)
    labels = results.pandas().xyxy[0]['name'].tolist()
    detected = set(labels)

    output = {}

    if 'person' in detected:
        output['human'] = True
    animal_found = [a for a in labels if a in animals]
    plant_found = [p for p in labels if p in plants]
    insect_found = [i for i in labels if i in insects]
    objects_found = [o for o in labels if o not in (animals + plants + insects + people)]

    if animal_found:
        output['animals'] = list(set(animal_found))
    if plant_found:
        output['plants'] = list(set(plant_found))
    if insect_found:
        output['insects'] = list(set(insect_found))
    if objects_found:
        output['objects'] = list(set(objects_found))

    return output, results

def save_image_for_training(img_pil, label):
    train_dir = f'data/train/{label}/'
    os.makedirs(train_dir, exist_ok=True)
    img_pil.save(os.path.join(train_dir, f'{label}_{len(os.listdir(train_dir))}.jpg'))

def retrain_yolo():
    data_yaml = 'data.yaml'
    weights_path = 'custom_yolo/best.pt'

    command = [
        'python', 'yolov5/train.py',
        '--img', '640',
        '--batch', '16',
        '--epochs', '10',
        '--data', data_yaml,
        '--weights', weights_path,
        '--name', 'intellidetect-update'
    ]

    subprocess.run(command)

    shutil.copy('yolov5/runs/train/intellidetect-update/weights/best.pt', 'custom_yolo/best.pt')

st.title("🧠 IntelliDetect: Smart Image Detector")
uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

if uploaded_file:
    with open("photo.jpg", "wb") as f:
        f.write(uploaded_file.read())
    st.image("photo.jpg", caption="Uploaded Image", use_column_width=True)

    img_cv2 = cv2.imread("photo.jpg")
    img_pil = Image.fromarray(cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB))

    output, results = run_detection("photo.jpg")

    st.subheader("Results")
    if output.get("human"):
        st.write("✅ Human detected.")
    if "animals" in output:
        st.write("🦁 Animal(s):", ', '.join(output["animals"]).title())
    if "plants" in output:
        st.write("🌿 Plant detected.")
    if "insects" in output:
        st.write("🐞 Insect detected.")
    if "objects" in output:
        st.write("📦 Objects:", ', '.join(output["objects"]).title())
    if not output:
        st.write("❌ Nothing recognizable detected.")

    # User verification for self-training
    label = st.selectbox("Confirm or label this image for training:", ['skip', 'human'] + animals + plants + insects + output.get("objects", []))
    if label != 'skip':
        save_image_for_training(img_pil, label)
        st.success(f"✅ Image saved under label: {label}")

        if st.button("🔁 Retrain Model Now"):
            st.info("Training in progress...")
            retrain_yolo()
            st.success("✅ Model retrained and updated!")
```

```python
import cv2
from deepface import DeepFace
from ultralytics import YOLO
import random
from google.colab.patches import cv2_imshow
img = cv2.imread("photo.jpg")

model = YOLO("yolov8n.pt")

results = model.predict(img, conf=0.5, verbose=False)[0]

animal_emotions = ['Calm', 'Alert', 'Aggressive', 'Scared']

try:
    result = DeepFace.analyze(img, actions=['emotion'], enforce_detection=False)
    for person in result:
        emotion = person['dominant_emotion']
        region = person['region']
        x, y, w, h = region['x'], region['y'], region['w'], region['h']
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(img, f"Human: {emotion}", (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
except:
    pass

for box in results.boxes.data.tolist():
    x1, y1, x2, y2, score, cls_id = box
    label = results.names[int(cls_id)]

    if label.lower() in ['dog', 'cat', 'horse', 'cow', 'sheep', 'bird']:
        animal_emotion = random.choice(animal_emotions)
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(img, f"{label} ({animal_emotion})", (int(x1), int(y1) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
